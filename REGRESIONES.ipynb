```python
# Buscamos predecir la probabilidad de que un usuario de 
# facebook se inscriba a un curso de programacion
# que conocemos?
# 1.Edad
# 2.Sexo
#3.Educacion
# 4.Nacionalidad
#Supongamos que tenemos una base de datos con 1000 alumnos.
#Queremos simular el evento planteado
Edad=np.random.normal(25,7,1000)
plt.figure(figsize=(12,8))
plt.title('Simulacion de edad')
plt.xlabel('Observacion')
plt.ylabel('Edad')
plt.plot(Edad,'o')
plt.show()
```

```python
#Histograma
plt.figure(figsize=(12,8))
plt.xlabel('Observacion')
plt.title('Simulacion de edad')
plt.ylabel('Edad')
plt.hist(Edad,bins=4)
plt.show()
```
