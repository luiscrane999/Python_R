```python
#Simulacio de la tierra 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Earth
earth={
    'Africa':{
        'initial population':1.4,
        'share women':.51},
    'Europa':{
        'initial population':.749,
        'share women':.54},
    'America':{
        'initial population':1,
        'share women':.49},
    'Asia':{
        'initial population':4.7,
        'share women':.47},
    'Oceania':{
        'initial population':0.46,
        'share women':.56}    
}
```

```python
earth['America']['initial population']
```

```python
#Continent
class Continent:
    def __init__(self,inputs):
        self.name=inputs['name']   
        self.initial_population=inputs['initial population']
        self.share_women=inputs['share women']
```

```python
#Ejmplo con clase 1
a=Continent({
    'name':'America',
    'initial population':earth['America']['initial population'],
    'share women':earth['America']['share women']
})
```

```python
#automatizacion
for continent in earth.keys():
    earth[continent]['class']=Continent({
    'name':'America',
    'initial population':earth[continent]['initial population'],
    'share women':earth[continent]['share women']})
```

```python
#Add growth rate
for continent in earth.keys():
    earth[continent]['class'].growth_rate=earth[continent]['class'].share_women/5
```

```python
#Forecast growth
years2model=50
time=range(years2model+1)

for continent in earth.keys():
    forecast_pop=[]
    for t in time:
        y=earth[continent]['class'].initial_population*np.exp(t*earth[continent]['class'].growth_rate)
        forecast_pop.append(y)
    earth[continent]['class'].forecast=forecast_pop
```

```python
plt.figure(figsize=(10,8))
for continent in earth.keys():
    plt.plot(range(len(earth[continent]['class'].forecast)),earth[continent]['class'].forecast,label=continent)
plt.legend()
plt.show()
```
